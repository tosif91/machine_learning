```python
import requests
import json
from types import SimpleNamespace
from PIL import Image

url_main = 'https://api.pexels.com/v1/'
```

```python
response = requests.get(url_main + 'search?query=nature')
```

```python
json_obj = json.loads(response.content,object_hook=lambda d : SimpleNamespace(**d))
```

```python
json_obj.photos[0]
json_obj.photos[0].src
```

```python
l_alt = []
l_avg_color = []
l_height = []
l_id = []
l_phtograhpher = []
l_phtograhpher_id = []
l_phtograhpher_url = []
```

```python
for i in range(len(json_obj.photos)):
  l_alt.append(json_obj.photos[i].alt)
  l_avg_color.append(json_obj.photos[i].avg_color)
  l_height.append(json_obj.photos[i].height)
  l_id.append(json_obj.photos[i].id)
  l_phtograhpher.append(json_obj.photos[i].photographer)
  l_phtograhpher_id.append(json_obj.photos[i].photographer_id)
  l_phtograhpher_url.append(json_obj.photos[i].photographer_url)
```

```python
l_landscape = []
l_large = []
l_large2x = []
l_medium = []
l_original = []
l_portrait = []
l_small = []
l_tiny = []
```

```python
for j in range(len(json_obj.photos)):
  l_landscape.append(json_obj.photos[j].src.landscape)
  l_large.append(json_obj.photos[j].src.large)
  l_large2x.append(json_obj.photos[j].src.large2x)
  l_medium.append(json_obj.photos[j].src.medium)
  l_original.append(json_obj.photos[j].src.original)
  l_portrait.append(json_obj.photos[j].src.portrait)
  l_small.append(json_obj.photos[j].src.small)
  l_tiny.append(json_obj.photos[j].src.tiny)
```

```python
import pandas as pd

photos_details_DF = pd.DataFrame({
    'Alt text':l_alt,
    'Avg Color':l_avg_color,
    'Height':l_height,
    'Id':l_id,
    'Photographer Name':l_phtograhpher,
    'Photographer Id':l_phtograhpher_id,
    'Photoprapher Url':l_phtograhpher_url,
    'Landscape Url':l_landscape,
    'Large img Url':l_large,
    'Large2x img Url':l_large2x,
    'Medium img Url':l_medium,
    'Original img Url':l_original,
    'Portrait img Url':l_portrait,
    'Small img Url':l_small,
    'Tiny img Url':l_tiny,

})
```

```python
photos_details_DF.to_excel('photos details from json api 2.xlsx')
```
