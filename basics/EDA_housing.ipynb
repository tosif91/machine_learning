```python
col_todrop = df.columns[   ((df.isnull().sum() / df.shape[0]) * 100  > 70).values ]
df.drop(columns=col_todrop,inplace=True)
print(col_todrop)
```

```python
df.LotFrontage =  df.LotFrontage.fillna(0)
df.MasVnrType.fillna('No Masonary',inplace=True)
df.MasVnrArea.fillna(0,inplace=True)
df.BsmtQual.fillna('No Basement',inplace=True)
df.BsmtExposure.fillna('No Basement',inplace=True)
df.BsmtCond.fillna('No Basement',inplace=True)
df.BsmtFinType1.fillna('No Basement',inplace=True)
df.BsmtFinType2.fillna('No Basement',inplace=True)
df.Electrical.dropna(inplace=True)
df.FireplaceQu.fillna('No FirePlace',inplace=True)
df.Fireplaces.fillna('No Fireplace',inplace=True)
df.GarageArea.fillna(0,inplace=True)
df.GarageCars.fillna(0,inplace=True)
df.GarageCond.fillna('Ng',inplace=True)
df.GarageType.fillna('Ng',inplace=True)
df.GarageFinish.fillna('Ng',inplace=True)
df.GarageYrBlt.fillna(0,inplace=True) 
df.GarageQual.fillna('Ng',inplace=True) 
df.drop(columns=['Electrical','Id'],inplace=True)
df.columns = df.columns.str.lower()
```
